```python
#Plotting Mathematical Functions

#Let’s plot a sine wave using NumPy.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)   # 100 points from 0 to 2π,  np.linspace(start, stop, num): Creates evenly spaced numbers.
y = np.sin(x)                      # sine of each x

plt.plot(x, y)                     # plot the sine curve
#plt.title(), plt.xlabel(), plt.ylabel(): Add descriptive labels.
plt.title("Sine Wave")            # add title
plt.xlabel("x (radians)")         # x-axis label
plt.ylabel("sin(x)")              # y-axis label
plt.grid(True)                    # add grid,  Makes the plot easier to read.
plt.show()
```

```python
#Cosine Wave Plot
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 2*np.pi, 100)   # 100 points from 0 to 2π
y = np.cos(x)                      # Cosine values

# Plotting
plt.plot(x, y)                     # Line plot of cosine wave
plt.title("Cosine Wave")          # Title of the plot
plt.xlabel("x (radians)")         # X-axis label
plt.ylabel("cos(x)")              # Y-axis label
plt.grid(True)                    # Show grid for better readability
plt.show()                        # Display the plot
```

# Plot Customization

```python
#Title, Axis Labels, and Legend
'''
label='...'	               Names the line (used in legend())
plt.legend()	           Displays a box to identify each line
plt.title()	               Adds a title above the plot
plt.xlabel() / ylabel()	   Add names to the axes
'''
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sine')   # Label for sine curve
plt.plot(x, y2, label='Cosine') # Label for cosine curve

plt.title("Sine and Cosine Waves")        # Plot title
plt.xlabel("x (radians)")                 # Label x-axis
plt.ylabel("Value")                       # Label y-axis
plt.legend()                              # Show legend box
plt.grid(True)
plt.show()
```

```python
#Line Color, Style, and Markers
'''
| Parameter   | Description                                   | Example          |
| ----------- | --------------------------------------------- | ---------------- |
| `color`     | Line color (`'red'`, `'blue'`, etc.)          | `color='red'`    |
| `linestyle` | Line style (`'-'`, `'--'`, `':'`)             | `linestyle='--'` |
| `marker`    | Data point marker (`'o'`, `'x'`, `'*'`, etc.) | `marker='o'`     |
'''
plt.plot(x, y1, color='red', linestyle='--', marker='o', label='Sine')
plt.plot(x, y2, color='blue', linestyle='-', marker='x', label='Cosine')
```

```python
#Axis Limits and Ticks
'''
| Function           | Description                            |
| ------------------ | -------------------------------------- |
| `xlim()`, `ylim()` | Set axis value range                   |
| `xticks()`         | Custom labels and positions for x-axis |

'''
plt.xlim(0, 2*np.pi)       # Set x-axis range
plt.ylim(-1.5, 1.5)        # Set y-axis range

plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', 'π/2', 'π', '3π/2', '2π'])  # Custom tick marks
```

```python
#Figure Size and Resolution
'''
figsize=(width, height) is in inches
dpi (dots per inch) controls sharpness of saved figures
'''
plt.figure(figsize=(8, 4), dpi=100)  # Width=8, Height=4, Resolution=100 dpi
```

```python
#All Together Example:
plt.figure(figsize=(8, 4), dpi=100)

plt.plot(x, y1, color='green', linestyle='--', marker='o', label='Sine')
plt.plot(x, y2, color='orange', linestyle='-', marker='x', label='Cosine')

plt.title("Sine and Cosine Waves")
plt.xlabel("x (radians)")
plt.ylabel("Amplitude")
plt.xlim(0, 2*np.pi)
plt.ylim(-1.5, 1.5)
plt.xticks([0, np.pi, 2*np.pi], ['0', 'π', '2π'])
plt.legend()
plt.grid(True)
plt.show()
```

# Multiple Plots

```python
#Multiple Lines on a Single Plot (Useful when comparing two related trends.)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sine', color='blue')
plt.plot(x, y2, label='Cosine', color='green')

plt.title("Sine and Cosine on Same Plot")
plt.xlabel("x (radians)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Using plt.subplot() (Index-based Subplots)  (Good for quick stacked comparisons.)
'''
| Function               | Description                           |
| ---------------------- | ------------------------------------- |
| `plt.subplot(r, c, i)` | r = rows, c = columns, i = plot index |
| `tight_layout()`       | Adjusts spacing to avoid text cutoff  |
'''
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  # (rows, cols, index) => First plot
plt.plot(x, y1, 'r')     #Color (e.g., 'r', 'g', 'b', 'k' for red, green, blue, black)
plt.title("Sine Wave")

plt.subplot(2, 1, 2)  # Second plot
plt.plot(x, y2, 'b')
plt.title("Cosine Wave")

plt.tight_layout()  # Prevent overlap
plt.show()
```

```python
#Using plt.subplots() (Recommended, Object-Oriented)  [ Best for clean, readable multi-panel plots.]
'''
| Element        | Description                                     |
| -------------- | ----------------------------------------------- |
| `fig, ax`      | `fig` is the whole figure, `ax` is each subplot |
| `ax[0].plot()` | Plot on the first subplot                       |
| `figsize`      | Control size for better layout                  |
'''
fig, ax = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

x = np.linspace(0, 2*np.pi, 100)
ax[0].plot(x, np.sin(x), color='purple')
ax[0].set_title("Sine")

ax[1].plot(x, np.cos(x), color='orange')
ax[1].set_title("Cosine")

plt.tight_layout()
plt.show()
```

```python
#Looping for Multiple Subplots [Useful for automated plotting of many graphs (sensor readings, experiments, etc.)]
'''
| Tool             | Use for                                |
| ---------------- | -------------------------------------- |
| `plt.plot()`     | Multiple lines in one graph            |
| `plt.subplot()`  | Quick subplot layout (index-based)     |
| `plt.subplots()` | Clean subplot layout (object-oriented) |
| `tight_layout()` | Prevent text/label overlapping         |
'''
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
'''
This line creates 4 subplots arranged in 2 rows × 2 columns.
figsize=(8, 6) sets the overall figure size in inches.
axs is a 2D array-like object where each item is one plot area (axis).
'''

functions = [np.sin, np.cos, np.tan, np.exp]           #functions: A list of mathematical functions.
titles = ['Sine', 'Cosine', 'Tangent', 'Exponential']  #titles: Corresponding titles for each plot.

x = np.linspace(0, 2, 100) #This creates 100 evenly spaced numbers from 0 to 2.Used as x-axis values for all functions.

'''
This nested loop goes through all 4 subplots.
idx calculates the index to pick the correct function and title.
Each subplot (axs[i][j]) plots one function with the right title
'''
for i in range(2):
    for j in range(2):
        idx = 2 * i + j
        axs[i][j].plot(x, functions[idx](x))
        axs[i][j].set_title(titles[idx])
'''
tight_layout() prevents overlap between subplots.
show() displays the final figure.
'''
plt.tight_layout()
plt.show()
```

# Bar, Histogram, and Pie Charts

```python
#Bar Chart (plt.bar())  [Used to compare discrete categories.]
'''
plt.bar(x, height): Makes vertical bars.
color='skyblue': Sets bar color.
grid(axis='y'): Adds horizontal grid lines only.
'''
import matplotlib.pyplot as plt

categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [10, 25, 15, 30]

plt.bar(categories, values, color='skyblue')
plt.title("Fruit Sales")
plt.xlabel("Fruits")
plt.ylabel("Quantity Sold")
plt.grid(axis='y')
plt.show()
```

```python
#Horizontal Bar Chart (plt.barh())  [Good for long category names or horizontal comparisons.]
plt.barh(categories, values, color='lightgreen')
plt.title("Fruit Sales (Horizontal)")
plt.xlabel("Quantity Sold")
plt.ylabel("Fruits")
plt.grid(axis='x')
plt.show()
```

```python
#Histogram (plt.hist())   [Used to show data distribution (like a frequency chart).]
'''
bins=20: Breaks the data into 20 intervals.
edgecolor: Adds visible borders between bars.
Good for sensor logs, test results, or probabilities.
'''
import numpy as np

data = np.random.randn(1000)  # 1000 random numbers ~ normal distribution

plt.hist(data, bins=20, color='salmon', edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#Pie Chart (plt.pie())   [Used to show proportions of a whole.]
'''
autopct='%1.1f%%': Shows percentages.
startangle=90: Rotates the chart to start from top.
axis('equal'): Ensures it’s not oval-shaped.
'''
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Market Share")
plt.axis('equal')  # Equal aspect ratio for a circle
plt.show()
```

# Scatter, Bubble, and Error Plots

```python
#Scatter Plot (plt.scatter())  [Used to display the relationship between two numeric variables.]
'''
plt.scatter(x, y): Each point (x, y) is a dot.
Good for trends, clusters, or anomalies.
'''
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = 2 * x + np.random.randn(50) * 0.1  # y ≈ 2x with some noise

plt.scatter(x, y, color='purple', marker='x')
plt.title("Scatter Plot: X vs Y")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

```python
#Bubble Chart (Size-based Scatter)  [Use case: IoT devices with size = signal strength or power usage.]
'''
s=: Size of each point (bubble).
alpha=0.5: Makes bubbles semi-transparent.
'''

x = np.random.rand(30)        #np.random.rand(30) generates 30 random values between 0 and 1
y = np.random.rand(30)
sizes = 500 * np.random.rand(30)  # Bubble sizes
'''
This creates 30 random values between 0 and 1, multiplied by 500.
So the sizes range from 0 to ~500 (area of each bubble).
Bigger number = bigger bubble.
'''

plt.scatter(x, y, s=sizes, alpha=0.5, color='green')
plt.title("Bubble Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
#Error Bars (plt.errorbar()) Shows data value with uncertainty range (±error).
'''
yerr=: Error values for each point
fmt='o': Circular markers
capsize=: End lines on error bars
Use case: Sensor readings with possible inaccuracy range.
'''
x = np.arange(1, 6)
y = np.array([2.0, 2.5, 3.7, 3.0, 4.1])
error = np.array([0.2, 0.3, 0.4, 0.1, 0.3])

plt.errorbar(x, y, yerr=error, fmt='o', capsize=5, color='red')
plt.title("Measurement with Error Bars")
plt.xlabel("Trial")
plt.ylabel("Measured Value")
plt.grid(True)
plt.show()
```

'''
Summary of Use Cases:

| Bubble Chart | `plt.scatter(s=)` | Showing a third variable (size)     |

# Advanced Layouts and Subplots
'''
Why this matters:

```python
#Subplots (plt.subplot() and plt.subplots()) [ Best when you want quick layout control via simple indexing.]

#A: plt.subplot(rows, cols, index)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(x, np.sin(x), 'r')
plt.title("Sine")

plt.subplot(1, 2, 2)  # Second subplot
plt.plot(x, np.cos(x), 'b')
plt.title("Cosine")

plt.tight_layout()
plt.show()
```

```python
#B: fig, ax = plt.subplots()  [More flexible, especially for advanced grid layouts.]
fig, ax = plt.subplots(2, 1, figsize=(8, 6))  # 2 rows, 1 column

x = np.linspace(0, 10, 100)

ax[0].plot(x, np.sin(x), 'g')
ax[0].set_title("Sine Wave")

ax[1].plot(x, np.cos(x), 'm')
ax[1].set_title("Cosine Wave")

plt.tight_layout()
plt.show()
```

```python
#Shared Axes [Useful when comparing functions with the same domain but different values.]
fig, ax = plt.subplots(2, 1, sharex=True, figsize=(8, 5))
'''
2, 1: Means 2 rows, 1 column (vertical layout).
sharex=True: Both plots use the same X-axis scale.
figsize=(8, 5): Sets the size of the whole figure (width=8, height=5).
ax: A list with 2 plot areas — ax[0] for the top, ax[1] for the bottom.
'''

x = np.linspace(0, 10, 100)  #100 values from 0 to 10.
ax[0].plot(x, np.sin(x))   #Top plot shows sin(x)
ax[1].plot(x, np.sin(2 * x))  #Bottom plot shows sin(2x) (a faster wave).

ax[0].set_title("Shared X Axis - sin(x) and sin(2x)")
plt.xlabel("X Axis shared")
plt.tight_layout()
plt.show()
```

```python
#Twin Axes (twinx()) [Used to show different data scales on the same plot.]
# Ideal when plotting things like temperature vs. power, or CPU usage vs. voltage.
fig, ax1 = plt.subplots()
#Creates a figure and one set of axes: ax1.This is the left Y-axis. 

'''
x: 100 values from 0 to 10.
y1: sine of x (values from -1 to 1)
y2: 100 × e^(-x), which decreases quickly from 100 to ~0
'''
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = 100 * np.exp(-x)

#Plots sine wave on ax1 in blue ('b').Left Y-axis is labeled "Sine" and colored blue
ax1.plot(x, y1, 'b')                        # Blue sine curve
ax1.set_ylabel("Sine", color='b')          # Left Y-axis label

#Creates a second Y-axis (right side), sharing the same X-axis.twinx() means “same X, new Y”.
ax2 = ax1.twinx()
ax2.plot(x, y2, 'r')                        # Red exponential curve, Plots the decaying curve in red ('r') on ax2
ax2.set_ylabel("Exponential Decay", color='r')   # Right Y-axis label, Right Y-axis is labeled "Exponential Decay" and colored red

plt.title("Dual Y-Axes: Sine & Exponential")
plt.grid(True)
plt.show()
```

```python
#Inset Plot (Zoomed View)  [ For zooming in on key data regions.]

import numpy as np
import matplotlib.pyplot as plt

# Create main plot area
fig, ax = plt.subplots()

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot sine wave on main axes
ax.plot(x, y)
ax.set_title("Main Plot with Inset")

# Add an inset plot (small plot inside the main one)
# [left, bottom, width, height] — all values are between 0 and 1 (relative to the figure)
inset = fig.add_axes([0.6, 0.5, 0.3, 0.3])  

# Plot the same data in the inset
inset.plot(x, y)

# Zoom into a specific part of the sine wave
inset.set_xlim(1, 2)       # Zoom x-axis from 1 to 2
inset.set_ylim(0.8, 1.0)   # Zoom y-axis to show sine peak
inset.set_title("Zoom In", fontsize=8)  # Smaller title for inset

# Show the full figure
plt.show()
```

# 3D Plots using mpl_toolkits.mplot3d

```python
#Setup: Import 3D Toolkit  [You must create a 3D subplot with projection='3d'.]
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
#3D Line Plot  [Visualize curves in space, like signals or trajectories.]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

z = np.linspace(0, 10, 100)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z, 'purple')
ax.set_title("3D Line Plot")
plt.show()
```

```python
#3D Scatter Plot  [Useful in sensor data, clusters, or spatial data (like IoT).]
import numpy as np
import matplotlib.pyplot as plt

# Create a new figure
fig = plt.figure()

# Add a 3D subplot (3D axes)
ax = fig.add_subplot(111, projection='3d')

# Generate 100 random points for x, y, z (all between 0 and 1)
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Plot a 3D scatter plot
# Points are colored based on their z value, using the 'plasma' color map
ax.scatter3D(x, y, z, c=z, cmap='plasma')

# Add a title
ax.set_title("3D Scatter Plot")

# Show the plot
plt.show()
```

```python
#3D Surface Plot  [Great for visualizing functions, terrain, or heat maps.]
'''
Explanation:

You create a grid of points on the XY plane using np.meshgrid.
For each (X, Y) point, you calculate Z as the sine of the distance from the origin.
plot_surface draws a 3D surface where height = Z.
The cmap='viridis' colors the surface nicely based on height.
Result: a smooth 3D wavy surface shaped like ripples spreading out from the center.
'''
import numpy as np
import matplotlib.pyplot as plt

# Create a new figure
fig = plt.figure()

# Add a 3D subplot (3D axes)
ax = fig.add_subplot(111, projection='3d')

# Create a grid of points in X and Y (from -5 to 5, 50 points each)
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)  # Make a 2D grid from X and Y vectors

# Calculate Z values based on X and Y (height values)
# Here, Z = sin(distance from origin)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the 3D surface using X, Y, and Z values
ax.plot_surface(X, Y, Z, cmap='viridis')  # Use the 'viridis' color map

# Add a title
ax.set_title("3D Surface Plot")

# Show the plot
plt.show()
```

```python
#Wireframe Plot
'''
Explanation:

Instead of a solid surface, plot_wireframe draws just the grid lines over the 3D surface.
Color of lines is set to green.
This helps you see the structure of the surface clearly.
Wireframe looks like a 3D mesh or net showing the shape of the surface without filling it.
'''
import numpy as np
import matplotlib.pyplot as plt

# Assuming X, Y, Z are already defined like this:
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a new figure
fig = plt.figure()

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot a wireframe (mesh grid lines) instead of a solid surface
ax.plot_wireframe(X, Y, Z, color='green')

# Add a title
ax.set_title("3D Wireframe Plot")

# Show the plot
plt.show()
```

'''

| Wireframe  | `ax.plot_wireframe()` | Structure/mesh visualization    |
